__üß¨ 1: Run the Training Code__

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
```

```python
# 1. PARAMETERS
IMG_SIZE = (224, 224) # Standard size for most Neural Networks
BATCH_SIZE = 32
```

```python
# 2. DATA LOADERS (Getting images into the model)
# We rescale pixels from 0-255 to 0-1 so the math is faster.
train_gen = ImageDataGenerator(rescale=1./255, horizontal_flip=True).flow_from_directory(
    'data/train', target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='binary'
)

val_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    'data/val', target_size=IMG_SIZE, batch_size=BATCH_SIZE, class_mode='binary'
)
```

```python
# 3. THE BRAIN: Transfer Learning
# We use MobileNetV2 because it is fast and "light" for cloud deployment.
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3), include_top=False, weights='imagenet'
)
base_model.trainable = False  # Freeze the pre-trained weights
```

```python
# 4. BUILDING THE MODEL
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1, activation='sigmoid') # Binary output: 0 or 1 (Pneumonia vs Normal)
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 5. TRAINING
print("Starting training... this will take about 5-10 minutes.")
model.fit(train_gen, validation_data=val_gen, epochs=5)
```

```python
# 6. SAVE THE MODEL
model.save('pneumonia_model.h5')
print("Model saved successfully as pneumonia_model.h5")
```

__üß™ 2: Test the Brain (Inference)__

```python
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
# 1. Pick a random image from the test set
# (Change the path to try different images)
img_path = 'data/test/PNEUMONIA/person100_bacteria_477.jpeg'
```

```python
# 2. Load and Preprocess the image
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) # Make it a "batch" of 1
img_array /= 255. # Normalize pixels
```

```python
# 3. Predict!
prediction = model.predict(img_array)
```

```python
# 4. Show Result
if prediction[0] > 0.5:
    print(f"Result: PNEUMONIA DETECTED (Confidence: {prediction[0][0]*100:.2f}%)")
else:
    print(f"Result: NORMAL LUNGS (Confidence: {(1-prediction[0][0])*100:.2f}%)")
```

```python
import matplotlib.pyplot as plt

# After running the prediction code:
plt.imshow(img)
plt.title(f"Prediction: {'Pneumonia' if prediction[0] > 0.5 else 'Normal'}")
plt.axis('off')
plt.show()
```

Verification Report:--->

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# 1. Get all images from the test folder
test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    'data/test', target_size=(224, 224), batch_size=32, class_mode='binary', shuffle=False
)

# 2. Make predictions for the whole test set
preds = model.predict(test_gen)
y_pred = (preds > 0.5).astype(int)
y_true = test_gen.classes

# 3. Print the report
print("--- Classification Report ---")
print(classification_report(y_true, y_pred, target_names=test_gen.class_indices.keys()))

# 4. Print the Confusion Matrix
print("--- Confusion Matrix ---")
print(confusion_matrix(y_true, y_pred))
```

________________________

`pneumonia_model.h5` is a "Heavy" format. It includes the model structure, the weights, and the optimizer states. __AWS Lambda__ has strict memory limits and needs to start up very fast. <br>
The Solution: __TensorFlow Lite__ (`.tflite`)
We will convert your model to TF Lite. This is like "dehydrating" the model‚Äîit removes all the unnecessary parts used for training and leaves only the parts needed for Inference (prediction). <br>
"Dehydrate" your model.

__üõ†Ô∏è 3: The Conversion Script__

```python
import tensorflow as tf

# 1. Load the heavy model
model = tf.keras.models.load_model('pneumonia_model.h5')

# 2. Initialize the converter
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# 3. Convert the model
tflite_model = converter.convert()

# 4. Save the "Light" model
with open('pneumonia_model.tflite', 'wb') as f:
    f.write(tflite_model)

print("Success! Your model is now dehydrated into 'pneumonia_model.tflite'")
```
